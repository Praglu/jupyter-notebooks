## Matplotlib basics

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Exercise 1

```python
x = np.linspace(0,2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y, marker=",", linestyle="-", label="$sin(x)$")
plt.axhline(y=1, marker="o", color="red", linestyle="--")
plt.axhline(y=-1, marker="o", color="red", linestyle="--")
plt.xlim(0,2*np.pi)
plt.yticks([-1, 0, 1])
plt.xlabel("$x$", fontsize=20)
plt.ylabel("$y$", fontsize=20, rotation=2)
plt.title("Wykres $sin(x)$", fontsize=20)
plt.legend(loc="center right")
plt.grid()
plt.show()
```

#### Exercise 3

```python
fig, axs = plt.subplots(nrows=2, ncols=2)

axs = axs.ravel()

x = np.linspace(0,2*np.pi,100)
y = [np.sin(x), np.cos(x), np.tan(x), 1.0/np.tan(x)]

titles = ["$\sin x$", "$\cos x$", "$tan x$", "$\cot x$"]

for i in range(4):
    axs[i].plot(x, y[i])
    axs[i].set_xlabel("$x$", fontsize=15)
    axs[i].set_ylabel("$y$", fontsize=15, rotation=2)
    axs[i].set_title(titles[i], fontsize=15)
    axs[i].set_xlim(0,2*np.pi)
    axs[i].grid()
    
plt.tight_layout()
plt.show()
```

#### Exercise 4

```python
data = {'matematyka': {'kolokwium': [4, 5, 3.5], 'egzamin': [3.5, 5, 4]}} 

labels_x = ["Student 1", "Student 2", "Student 3"]
kolokwium = data['matematyka']['kolokwium']
egzamin = data['matematyka']['egzamin']

x = np.arange(len(labels_x))
bar_width = 0.25
bar_height = np.arange(5)

fig, ax = plt.subplots()
first_bar = ax.bar(x-bar_width/2, kolokwium, bar_width, label="Kolokwium")
second_bar = ax.bar(x+bar_width/2, egzamin, bar_width, label="Egazamin")

ax.set_ylabel("Oceny")
ax.set_title("Wyniki z matematyki")
ax.set_xticks(x)
ax.set_xticklabels(labels_x)
ax.legend(loc="best")

fig.tight_layout()
plt.show()
```

#### Exercise 5

```python
import random

def function(x):
    return np.sin(x)

x = np.arange(0,10,0.1)
y = function(x)

plt.plot(x, y,'k-', lw=2)
xp = [random.uniform(0,np.pi) for _ in range(500)]
yp = [random.random() for _ in range(500)]
new_yp = []
new_xp = []
not_in_yp = []
not_in_xp = []

for i in range(len(yp)):
    if yp[i] - function(xp[i]) <= 0:
        new_yp.append(yp[i])
        new_xp.append(xp[i])
    elif yp[i] - function(xp[i]) > 0:
        not_in_yp.append(yp[i])
        not_in_xp.append(xp[i])
    else:
        continue

plt.scatter(new_xp,new_yp, color="red", marker=".")
plt.scatter(not_in_xp,not_in_yp, color="blue", marker=".")
plt.xlabel("$x$")
plt.ylabel("$\sin x$")

plt.xlim(0,np.pi)
plt.ylim(0,1.1)

plt.show()
```
